```python
from numba import jit

import math
import librosa
import librosa.display

import IPython.display as ipd

import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
from scipy.signal import chirp, spectrogram, stft
import matplotlib.pyplot as plt

import numpy as np
import pandas as pd

plt.rcParams['axes.labelsize']  = 16
plt.rcParams['axes.titlesize']  = 16
plt.rcParams['legend.fontsize'] = 16
plt.rcParams['xtick.labelsize'] = 16
plt.rcParams['ytick.labelsize'] = 16
plt.rcParams['lines.linewidth'] = 2
plt.rcParams["figure.figsize"] = (20,10)

plt.ion()
```

```python
def plot_chirp_spectogram():
    t = np.linspace(0, 10, 1500)
    w = chirp(t, f0=1000, f1=10000, t1=100, method='linear')
    plt.plot(t, w)
    plt.title("Linear Chirp, f(0)=6, f(10)=1")
    plt.xlabel('t (sec)')
    plt
    plt.show()
    
def plot_spectrogram(title, w, fs):
    ff, tt, Sxx = spectrogram(w, fs=fs)
    plt.pcolormesh(tt, ff[:145], Sxx[:145], shading='gouraud')
    plt.title(title)
    plt.xlabel('t (sec)')
    plt.ylabel('Frequency (Hz)')
    plt.grid()
    
def plot_stft(title, w, fs):
    ff, tt, Sxx = stft(w, fs=fs)
    plt.pcolormesh(tt, ff, np.log(np.abs(Sxx)), shading='gouraud')
    plt.title(title)
    plt.xlabel('t (sec)')
    plt.ylabel('Frequency (Hz)')
    plt.grid()
    
    
fs = 22050
T = 100
t = np.arange(0, int(T*fs)) / fs
w = chirp(t, f0=1000, f1=10000, t1=100, method='linear')

# plot_chirp_spectogram()
plot_spectrogram('spectogram', w, fs)
# plot_stft('stft', w, fs)
```
